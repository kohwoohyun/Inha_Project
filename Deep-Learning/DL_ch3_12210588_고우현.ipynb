```python
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
    return np.array(x > 0, dtype = int)

x = np.arange(-7.0, 7.0, 0.1)
y = step_function(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def sigmond(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-7.0, 7.0, 0.1)
y = sigmond(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-7.0, 7.0, 0.1)
y = relu(x)
plt.plot(x,y)
plt.ylim(-1.5, 7.1)
plt.show()
```

```python
def init_network():
    network = {}
    network['W1'] = np.array([[0.4, 0.3, 0.6], [0.1, 0.5, 0.2]])
    network['B1'] = np.array([1.1, 1.4, 1.3])
    network['W2'] = np.array([[0.3, 0.4], [0.8, 0.5], [0.4, 0.1]])
    network['B2'] = np.array([0.9, 0.1])
    network['W3'] = np.array([[1.1, 0.9], [0.7, 0.8]])
    network['B3'] = np.array([0.1, 0.3])
    return network

def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    B1, B2, B3 = network['B1'], network['B2'], network['B3']
    A1 = np.dot(x, W1) + B1
    Z1 = sigmond(A1)
    A2 = np.dot(Z1, W2) + B2
    Z2 = sigmond(A2)
    A3 = np.dot(Z2, W3) + B3
    Y = identity_function(A3)
    return Y

network = init_network()
x = np.array([3.0, 4.0])
y = forward(network, x)
print(y)
```

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y

a = np.array([0.3, 1.2, 2.5])
y = softmax(a)
print(y)
```

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y

test_array = np.array([0.1, 0.2, 0.3])
test_array2 = np.array([1000, 2000, 3000])
y1 = softmax(test_array)
y2 = softmax(test_array2)
print(y1)
print(y2)
```
