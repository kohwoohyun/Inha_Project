# NAND 게이트만으로 컴퓨터를 만들 수 있다.Why?

NAND 게이트로 AND, OR 게이트를 구현할 수 있고, 이를 통해 다층 퍼셉트론으로 XOR 게이트를 구현할 수 있다. 따라서, NAND 게이트를 이용해 선형적인 표현 뿐만 아니라 비선형적인 표현을 구현할 수 있고, 컴퓨터가 수행하는 처리도 모두 표현이 가능하므로 컴퓨터를 만들 수 있다.

# AND 게이트

```python
def AND(x1, x2):
    w1, w2, theta = 0.6, 0.7, 0.9
    tmp = x1 * w1 + x2 * w2
    if tmp <= theta :
        return 0
    else :
        return 1

print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

# OR 게이트

```python
def OR(x1, x2):
    w1, w2, theta = 0.6, 0.7, 0.3
    tmp = x1 * w1 + x2 * w2
    if tmp <= theta :
        return 0
    else :
        return 1

print(OR(0, 0))
print(OR(0, 1))
print(OR(1, 0))
print(OR(1, 1))
```

# NAND 게이트

```python
def NAND(x1, x2):
    w1, w2, theta = -0.6, -0.7, -0.9
    tmp = x1 * w1 + x2 * w2
    if tmp <= theta :
        return 0
    else :
        return 1

print(NAND(0, 0))
print(NAND(0, 1))
print(NAND(1, 0))
print(NAND(1, 1))
```

# XOR 게이트

```python
def XOR(x1,x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

print(XOR(0, 0))
print(XOR(0, 1))
print(XOR(1, 0))
print(XOR(1, 1))
```

# NAND 게이트만으로 OR 게이트 구현

```python
def NEW_OR(x1, x2):
    s1 = NAND(x1, x1)
    s2 = NAND(x2, x2)
    y = NAND(s1, s2)
    return y

print(NEW_OR(0, 0))
print(NEW_OR(1, 0))
print(NEW_OR(0, 1))
print(NEW_OR(1, 1))
```
