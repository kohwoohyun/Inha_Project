# 과제 1

```python
import numpy as np

A = np.array([[1, 0, 2], [0, 1, 1], [1, 2, 2]])
E = np.identity(3)
# A*B = 2B - A
# A*B - 2B = -A
# (A - 2*E)*B = -A
C = A - 2*E
D = -A

B = np.dot(np.linalg.inv(C), D)
print(B)
```

# 과제 2-1

```python
import numpy as np

A = np.array([[2, -1, -3],
               [-1, -2, -3],
               [1, 1, 4]])
B = np. array([[2],
                [-1],
                [1]])

X = np.linalg.solve(A, B)
print(X)
```

# 과제 2-2

```python
import numpy as np

A = np.array([[2, -1, -3],
              [-4, 2, 6],
              [3, 1, 4]])
B = np.array([[2],
              [-1],
              [1]])

try:
    X = np.linalg.solve(A, B)
    print(X)
except np.linalg.LinAlgError as e:
    print("Error :",e)
```

# 과제 3

```python
import numpy as np

def f(array):
    n = len(array)
    A = []

    for i in range(n):
        for j in range(n):
            A.append(array[i][j])

    result = np.array(A)
    return result

f(np.array([[1,2],[2,3]]))
```

# 과제 4

```python
import numpy as np

A = np.array([[0.6, 0.3],
              [0.4, 0.7]])
init_vector = np.array([[0, 1]])
x0 = init_vector.T
x1 = A.dot(x0)
x2 = A.dot(x1)
print(x2)
```

# 과제 5

```python
import numpy as np

def x(t):
    A = np.array([[0.6, 0.3],
                  [0.4, 0.7]])
    init_vector = np.array([[1],
                            [0]])
    xt = init_vector
    
    if t == 0:
        return xt
    else:
        for i in range(t):
            xt = A.dot(xt)
        return xt

x(10)
```
