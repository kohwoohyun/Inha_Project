```python
np.random.seed(0)

x = 2*np.random.randn(100,1)
y = 8+5*x+np.random.randn(100,1)

plt.figure(figsize=(5,5))
plt.scatter(x,y)
plt.show()
```

```python
def get_cost(y,y_pred):
    N = len(y)
    cost = np.sum(np.square(y-y_pred))/N
    return cost
```

```python
def get_weight_update(w,b,x,y,learning_rate=0.01):
    N = len(y)

    w_update = np.zeros_like(w)
    b_update = np.zeros_like(b)

    y_pred = np.dot(x,w.T)+b
    diff = y-y_pred

    b_factors = np.ones((N,1))

    w_update = -(2/N)*learning_rate*(np.dot(x.T,diff))
    b_update = -(2/N)*learning_rate*(np.dot(b_factors.T,diff))

    return w_update, b_update
```

```python
def gradient_descent_steps(x,y,epochs):
    w = np.zeros((1,1))
    b = np.zeros((1,1))

    for _ in range(epochs):
        w_update, b_update = get_weight_update(w,b,x,y,learning_rate=0.01)
        w = w-w_update
        b = b-b_update

    return w, b
```

```python
w,b = gradient_descent_steps(x,y,epochs=2000)
print(w,b)
y_pred = w*x+b

print(get_cost(y,y_pred))
```

```python
# sklearn.datasets 중 fetch_california_housing 데이터 사용
from sklearn.datasets import fetch_california_housing

california = fetch_california_housing()

californiaDF = pd.DataFrame(california.data, columns=california.feature_names)
californiaDF['PRICE'] = california.target

print('California 데이터 세트 크기:', californiaDF.shape)
californiaDF.head()
```

```python
y_target = californiaDF['PRICE']
X_data = californiaDF.drop(['PRICE'], axis=1, inplace = False)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_data, y_target, test_size = 0.1, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet

linear = LinearRegression()
linear.fit(X_train, y_train)
y_preds_LR = linear.predict(X_test)

ridge = Ridge(alpha = 10)
ridge.fit(X_train, y_train)
y_preds_RIDGE = ridge.predict(X_test)

lasso = Lasso(alpha = 0.1)
lasso.fit(X_train, y_train)
y_preds_LASSO = lasso.predict(X_test)

elasticnet = ElasticNet(alpha = 0.5, l1_ratio = 0.5)
elasticnet.fit(X_train, y_train)
y_preds_EN = elasticnet.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

print('LinearRegression')

mae1 = mean_absolute_error(y_test, y_preds_LR)
mse1 = mean_squared_error(y_test, y_preds_LR)
rmse1 = np.sqrt(mse1)
r2_1 = r2_score(y_test, y_preds_LR)

print('Mean Absolute Error : {:.3f}'.format(mae1))
print('Mean Squared Error : {:.3f}'.format(mse1))
print('Root Mean Squared Error : {:.3f}'.format(rmse1))
print('Variance Score : {:.3f}'.format(r2_1))

print('***************************************************************')

print('Ridge')

mae2 = mean_absolute_error(y_test, y_preds_RIDGE)
mse2 = mean_squared_error(y_test, y_preds_RIDGE)
rmse2 = np.sqrt(mse2)
r2_2 = r2_score(y_test, y_preds_RIDGE)

print('Mean Absolute Error : {:.3f}'.format(mae2))
print('Mean Squared Error : {:.3f}'.format(mse2))
print('Root Mean Squared Error : {:.3f}'.format(rmse2))
print('Variance Score : {:.3f}'.format(r2_2))

print('***************************************************************')

print('Lasso')

mae3 = mean_absolute_error(y_test, y_preds_LASSO)
mse3 = mean_squared_error(y_test, y_preds_LASSO)
rmse3 = np.sqrt(mse3)
r2_3 = r2_score(y_test, y_preds_LASSO)

print('Mean Absolute Error : {:.3f}'.format(mae3))
print('Mean Squared Error : {:.3f}'.format(mse3))
print('Root Mean Squared Error : {:.3f}'.format(rmse3))
print('Variance Score : {:.3f}'.format(r2_3))

print('***************************************************************')

print('ElasticNet')

mae4 = mean_absolute_error(y_test, y_preds_EN)
mse4 = mean_squared_error(y_test, y_preds_EN)
rmse4 = np.sqrt(mse4)
r2_4 = r2_score(y_test, y_preds_EN)

print('Mean Absolute Error : {:.3f}'.format(mae4))
print('Mean Squared Error : {:.3f}'.format(mse4))
print('Root Mean Squared Error : {:.3f}'.format(rmse4))
print('Variance Score : {:.3f}'.format(r2_4))

print('***************************************************************')
```
