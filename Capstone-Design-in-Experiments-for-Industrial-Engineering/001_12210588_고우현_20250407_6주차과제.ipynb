# How to combine data from multiple tables

```python
# 인천국제공항공사 환승승객현황 정보 중 2024년 상반기 도착과 2024년 상반기 출발 이용

arrival = pd.read_csv('arrival.csv', parse_dates=True, encoding='euc-kr')
arrival
```

```python
arrival = arrival[['운항일자', '계획시간', '운항편명', '환승승객']]
arrival
```

```python
departure = pd.read_csv('departure.csv', parse_dates=True, encoding='euc-kr')
departure
```

```python
departure = departure[['운항일자', '계획시간', '운항편명', '환승승객']]
departure
```

```python
total = pd.concat([arrival, departure], axis=0)
total
```

```python
print('Shape of the "arrival" table: ', arrival.shape)
print()
print('Shape of the "departure" table: ', departure.shape)
print()
print('Shape of the resulting "total" table: ', total.shape)
```

```python
total = total.sort_values('운항일자')
total
```

```python
total_ = pd.concat([arrival, departure], keys=['미국','중국'])
total_
```

```python
# Join tables using a common identifier
# 인천국제공항공사 환승승객현황 정보 2024년 상반기 출발, 도착 이용

arrival = pd.read_csv('arrival.csv', encoding='euc-kr')
arrival
```

```python
departure = pd.read_csv('departure.csv', encoding='euc-kr')
departure
```

```python
_departure_ = pd.merge(departure, arrival, how='left', on='공항코드')
_departure_
```

```python
_arrival_ = pd.merge(arrival, departure, how='left', left_on='공항코드', right_on='출발국가')
_arrival_
```

# How to handle time series data with ease

```python
#인천국제공항공사 환승승객현황 정보 중 2024년 상반기 도착 정보 이용

arrival = pd.read_csv('arrival.csv', encoding='euc-kr')
arrival = arrival.rename(columns={'운항일자': 'datetime'})
arrival
```

```python
arrival['datetime']
```

```python
arrival['datetime'] = pd.to_datetime(arrival['datetime'])
arrival['datetime']
```

```python
arrival['datetime'].min(), arrival['datetime'].max()
```

```python
arrival['datetime'].max() - arrival['datetime'].min()
```

```python
arrival['month'] = arrival['datetime'].dt.month
arrival
```

```python
arrival.groupby([arrival['datetime'].dt.weekday, '도착국가'])['환승승객'].mean()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(12,4))
arrival.groupby([arrival['datetime'].dt.weekday, '도착국가'])['환승승객'].mean().plot(kind='bar', rot=0, ax=axs)
plt.xlabel('Weekday')
plt.ylabel('Passenger')
plt.show()
```

```python
country = arrival.pivot(index='datetime', columns='운항편명', values='환승승객')
country
```

```python
# 서울특별시 연구기관 현황 통계에서 일부 수정 후 이용

data = pd.read_csv('research.txt', sep='\t', header=4)

data.drop(['합계', '소계'], axis=1, inplace=True)

columns = ['People_' + str(i) for i in range(1,13)]
data.columns = columns

data.head()
```

```python
# Cycle

plt.figure(figsize=(10, 5))

data['People_1'].plot()

plt.title('People value according to index')

plt.show()
```

```python
plt.figure(figsize=(10, 5))

data.groupby('People_1')['People_3'].max().plot()

plt.title('Max People value according to index')
plt.xlabel('People_1')
plt.ylabel('People_2')

plt.show()
```

```python
# Op_setting

data[data['People_1']==170]
```

```python
for i in range(2, 5):
    plt.figure(figsize=(10, 5))
    data[data['People_1']==170]['People_'+str(i)].plot()

    plt.title('People according to People_1')
    plt.xlabel('People_1')
    plt.ylabel('Value of people')

    plt.show()
```

```python
data.reset_index(drop=True, inplace=True)

for col in columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
people_list = ['People_2', 'People_3']

for i in range(len(people_list)):
    plt.figure(figsize=(10,5))
    plt.hist(data['People_'+str(i+1)], bins=100)

    plt.title('People Histogram')

    plt.show()
```

```python
# People_2와  People_3 간의 상관관계

for i in range(len(people_list)):
    x = data[people_list[i]]

    for j in range(len(people_list)):
        if i == j:
            continue
        y = data[people_list[j]]

        plt.figure(figsize=(5, 5))

        plt.title(people_list[i] + ' vs. ' + people_list[j])
        plt.xlabel(people_list[i])
        plt.ylabel(people_list[j])

        plt.scatter(x, y)
        plt.show()
```
