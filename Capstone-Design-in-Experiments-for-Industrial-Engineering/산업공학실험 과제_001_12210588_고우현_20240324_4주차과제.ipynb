# 수정구 재외국민의 평균, 분산, 중앙값을 구하라

```python
print(data.iloc[0:17]['재외국민'].mean())
print(data.iloc[0:17]['재외국민'].var())
print(data.iloc[0:17]['재외국민'].median())
```

```python
# Column 산술 연산을 통해 19세 이상 인구수 구하기
data['19세 이상 인구수'] = data['19세 이상_남'] + data['19세 이상_여']
data
```

```python
# Column 산술 연산을 통해 19세 이상 남녀 인구수 차이 구하기
data['19세 이상 남녀 차이'] = data['19세 이상_남'] - data['19세 이상_여']
data
```

```python
# Column 산술 연산을 통해 재외국민과 세대수 비율 구하기
data['재외국민/세대수'] = data['재외국민'] / data['세대수']
data
```
