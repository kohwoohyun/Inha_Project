# How do I create plots in pandas?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# weather.csv에서 테마분류, 경도, 위도를 따로 뽑아내 새로운 csv파일을 생성

location = pd.read_csv("new_weather.csv", index_col=0)
location
```

```python
location.plot()
plt.show()
```

```python
location["longitude"].plot()
plt.show()
```

```python
location.plot.scatter(x="longitude", y="latitude", alpha=0.5)
plt.show()
```

```python
location.plot.box()
plt.show()
```

```python
axs = location.plot.area(figsize=(12, 4), subplots=True)
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12, 4))
location.plot.area(ax=axs)
axs.set_ylabel("location")
plt.show()
```

# Case Study

```python
# 안면도, 고산, 울릉도, 포항 총자외선지수 연도별 월평균

uvi = pd.read_csv('uvi_ym_avg_new.txt', sep=",", encoding="euc-kr")
uvi.head()
```

```python
uvi.drop("Unnamed: 5", axis=1, inplace=True)
uvi.head()
```

```python
index_column = ['Date']
place_columns = ['Place_'+str(i) for i in range(1,5)]

columns = index_column + place_columns
uvi.columns = columns
uvi.head()
```

```python
uvi[uvi['Date'] == 200110]
```

```python
uvi[uvi['Date'] == 201001].describe()
```

```python
uvi.groupby('Place_1')['Place_4'].mean()
```

```python
max_place = uvi.groupby('Date')['Place_2'].max().reset_index()
max_place.rename(columns={'Place_2':'MaxPlace'}, inplace=True)

max_place
```

```python
temp_df = pd.merge(uvi, max_place, on='Date', how='left')

uvi['RenewPlace'] = temp_df['MaxPlace']- uvi['Place_4']
uvi.head()
```
