# 11주차 내용

```python
# sklearn.datasets에서 load_wine 사용
from sklearn.datasets import load_wine

wine = load_wine()
```

```python
wine
```

```python
wine.keys()
```

```python
wine.DESCR
```

```python
X = wine.data
y = wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size = 0.2, random_state = 42)
```

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
print(f'test set accuracy: {model.score(X_test, y_test)}')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
print(f'test set accuracy: {model.score(X_test, y_test)}')
```

```python
import sklearn.svm as svm
```

```python
model = svm.SVC(kernel = 'linear', random_state=42)
model.fit(X_train, y_train)
```

```python
print(f'test set accuracy: {model.score(X_test, y_test)}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
print(f'test set accuracy: {model.score(X_test, y_test)}')
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('정확도(Accuracy): {:.4f}'.format(accuracy_score(y_test, pred)))
```

```python
from sklearn.metrics import precision_score, recall_score

print('정밀도(Precision): {:.4f}'.format(precision_score(y_test,pred, average='macro')))
print('재현율(Recall): {:.4f}'.format(recall_score(y_test,pred, average='macro')))
```

```python
from sklearn.metrics import f1_score

print('F1_score: {:.4f}'.format(f1_score(y_test,pred, average='macro')))
```
