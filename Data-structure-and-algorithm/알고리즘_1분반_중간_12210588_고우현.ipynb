# 1. 알고리즘 구현



## 1. 버블 정렬

```python
def bubble_sort(arr):
    n = len(arr)                                      # 리스트의 길이를 저장합니다.

    for i in range(n - 1):                            # n-1 번의 '패스'를 반복합니다. (가장 큰 요소가 끝으로 이동)
        for j in range(n - 1 - i):                    # 이미 정렬된 마지막 i개 요소를 제외하고 비교합니다.
            if arr[j] > arr[j+1]:                     # 현재 요소가 다음 요소보다 크면
                arr[j], arr[j+1] = arr[j+1], arr[j]   # 두 요소를 교환(swap)합니다.

    return arr
```

## 2. 삽입 정렬

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):                      # 두 번째 요소(인덱스 1)부터 시작하여 리스트를 순회합니다.
        key = arr[i]                                  # 현재 요소를 'key'로 설정합니다. (정렬될 요소)
        j = i - 1                                     # key의 왼쪽 인덱스를 'j'로 설정합니다.

        while j >= 0 and key < arr[j]:                # j가 0보다 크거나 같고, key보다 큰 요소가 왼쪽에 있는 동안
            arr[j + 1] = arr[j]                       # 그 요소를 오른쪽으로 한 칸 이동시킵니다.
            j -= 1                                    # 인덱스 j를 감소시켜 다음 왼쪽 요소로 이동합니다.

        arr[j + 1] = key                              # 반복이 멈춘 위치(j+1)에 key를 삽입합니다.

    return arr
```

## 3. 선택 정렬

```python
def selection_sort(arr):
    n = len(arr)                                          # 리스트의 길이를 저장합니다.

    for i in range(n - 1):                                # 리스트의 첫 번째 요소부터 n-2번째 요소까지 순회합니다.
        min_idx = i                                       # 현재 위치(i)를 가장 작은 요소의 인덱스로 가정합니다.

        for j in range(i + 1, n):                         # i+1부터 리스트 끝까지 순회하며 가장 작은 요소를 찾습니다.
            if arr[j] < arr[min_idx]:                     # 현재까지의 최소값보다 더 작은 요소를 찾으면
                min_idx = j                               # min_idx를 그 요소의 인덱스로 업데이트합니다.

        if min_idx != i:                                  # 현재 위치(i)와 찾은 최소값의 위치(min_idx)가 다르면
            arr[i], arr[min_idx] = arr[min_idx], arr[i]   # 두 요소를 교환합니다.

    return arr
```

## 4. 병합 정렬

```python
def merge_sort(arr):
    if len(arr) <= 1:                       # 리스트 길이가 1 이하이면 이미 정렬된 것으로 간주하고 반환합니다. (재귀 종료 조건)
        return arr

    mid = len(arr) // 2                     # 리스트를 반으로 나눕니다.
    left_half = arr[:mid]
    right_half = arr[mid:]

    left = merge_sort(left_half)            # 왼쪽과 오른쪽 절반을 재귀적으로 정렬합니다.
    right = merge_sort(right_half)

    return merge(left, right)               # 정렬된 두 리스트를 병합(merge)하는 함수를 호출합니다.

def merge(left, right):
    merged = []                             # 병합된 결과를 저장할 리스트를 초기화합니다.
    i = j = 0                               # 왼쪽과 오른쪽 리스트의 현재 인덱스를 0으로 초기화합니다.

    while i < len(left) and j < len(right): # 왼쪽 또는 오른쪽 리스트의 요소가 남아있는 동안 반복합니다.
        if left[i] <= right[j]:             # 왼쪽 요소가 오른쪽 요소보다 작거나 같으면
            merged.append(left[i])          # 왼쪽 요소를 merged 리스트에 추가하고 왼쪽 인덱스를 증가시킵니다.
            i += 1
        else:
            merged.append(right[j])         # 오른쪽 요소를 merged 리스트에 추가하고 오른쪽 인덱스를 증가시킵니다.
            j += 1

    merged.extend(left[i:])                 # 남아있는 왼쪽 리스트의 요소들을 모두 추가합니다.
    merged.extend(right[j:])                # 남아있는 오른쪽 리스트의 요소들을 모두 추가합니다.

    return merged                           # 병합된 최종 리스트를 반환합니다.
```

## 5. 퀵 정렬

```python
def quick_sort(arr):
    if len(arr) <= 1:                                           # 리스트 길이가 1 이하이면 이미 정렬된 것으로 간주하고 반환합니다. (재귀 종료 조건)
        return arr

    else:
        pivot = arr[0]                                          # 리스트의 첫 번째 요소를 피벗(pivot)으로 선택합니다.
        less = [x for x in arr[1:] if x <= pivot]               # 피벗보다 작은 요소들을 모은 리스트입니다.
        greater = [x for x in arr[1:] if x > pivot]             # 피벗보다 큰 요소들을 모은 리스트입니다.

        return quick_sort(less) + [pivot] + quick_sort(greater) # 작은 요소들을 재귀적으로 퀵 정렬하고, 피벗을 추가한 후, 큰 요소들을 재귀적으로 퀵 정렬하여 결합합니다.
```

# 2. 알고리즘 테스트 진행

```python
# 이전 1번 섹션에서 정의된 정렬 함수들을 사용합니다.
# (bubble_sort, insertion_sort, selection_sort, merge_sort, quick_sort)

test_data_original = [9, 3, 7, 2, 8, 1, 6, 5, 4]  # 원본 리스트를 나타내는 변수명
expected_result = sorted(test_data_original)

print("--- 알고리즘 테스트 진행 ---")
print(f"원본 데이터: {test_data_original}")
print(f"기대 결과: {expected_result}")
print("-" * 50)


def run_test_and_verify(name, func, original_data, expected_data):

    new_list2 = original_data.copy()  # 원본 보호를 위해 복사본을 생성합니다.

    sorted_data = func(new_list2) # 정렬 실행 (함수 내부에서 복사본을 한 번 더 생성할 수도 있지만, 여기서는 이미 복사된 리스트를 전달합니다.)

    is_correct = sorted_data == expected_data # 정확성 검증

    is_original_preserved = original_data == [9, 3, 7, 2, 8, 1, 6, 5, 4] # 원본 보호 검증 (test_data_original이 변경되지 않았는지 확인)

    print(f"[{name}] 테스트")
    print(f"  정렬 결과: {sorted_data}")
    print(f"  원본 상태: {original_data}")
    print(f"  정확성: {'성공' if is_correct else '실패'}")
    print(f"  원본 보호: {'성공' if is_original_preserved else '실패'}")
    print("-" * 50)


run_test_and_verify("Bubble Sort", bubble_sort, test_data_original, expected_result) # 버블 정렬
run_test_and_verify("Insertion Sort", insertion_sort, test_data_original, expected_result) # 삽입 정렬
run_test_and_verify("Selection Sort", selection_sort, test_data_original, expected_result) # 선택 정렬
run_test_and_verify("Merge Sort", merge_sort, test_data_original, expected_result) # 병합 정렬
run_test_and_verify("Quick Sort", quick_sort, test_data_original, expected_result) # 퀵 정렬
```

# 3. 데이터 생성

```python
import random

test_range = []
for exponent in range(1, 6):
    test_range.append(10 ** exponent)

MAX_SIZE = 100000
MAX_VALUE = 10000000
alist = random.sample(range(1, MAX_VALUE), MAX_SIZE)

print("--- 데이터 생성 ---")
print(f"테스트 범위: {test_range}")
print(f"원본 데이터 alist 크기: {len(alist):,}개")

print("-" * 50)
```

# 4. 데이터 크기별 실행 시간 측정

```python
import time
import numpy as np
import pandas as pd

# 측정 함수 목록 (딕셔너리)
ALGORITHMS_TIME = {
    "Bubble Sort": bubble_sort, "Insertion Sort": insertion_sort,
    "Selection Sort": selection_sort, "Merge Sort": merge_sort,
    "Quick Sort": quick_sort
}
DATA_SIZES = test_range # 3번 섹션에서 정의된 test_range 사용

# 딕셔너리 초기화: {알고리즘 이름: {데이터 크기: 실행 시간}}
timing_results = {name: {} for name in ALGORITHMS_TIME}

print("--- 데이터 크기별 실행 시간 측정 ---")


def time_calculation(sort_func, func_name, original_list, sizes):
    """
    주어진 정렬 함수(sort_func)에 대해 크기별 시간을 계산하고 출력합니다.
    """
    result = {}

    print(f"\n[알고리즘: {func_name}] 측정 시작...")

    for size in sizes:

        copy_list = original_list[:size]

        start_time = time.time()
        sort_func(copy_list)
        elapsed_time = time.time() - start_time

        result[size] = elapsed_time

        print(f"  > 크기 {size:,}: {elapsed_time} 초")


    return result

# 각 정렬 방식마다 시간 계산 함수 실행 및 결과 저장
for name, func in ALGORITHMS_TIME.items():
    timing_results[name] = time_calculation(func, name, alist, DATA_SIZES)

print("\n--- 실행 시간 측정 완료 ---")
print("-" * 50)
```

# 5. 그래프 생성

```python
import matplotlib.pyplot as plt

DATA_SIZES = [10, 100, 1000, 10000, 100000]
ALGORITHM_NAMES = list(timing_results.keys())


results_df = pd.DataFrame(timing_results).T

print("--- 그래프 생성 ---")

plt.figure(figsize=(12, 8))
plt.xlabel("# n")
plt.ylabel("seconds")

plt.yscale('log')
plt.xscale('log')

# Pandas DataFrame을 사용하여 그래프 그리기
for name in results_df.index:
    # results_df.columns는 데이터 크기(X축), results_df.loc[name].values는 실행 시간(Y축)입니다.
    plt.plot(results_df.columns, results_df.loc[name].values, marker='o', label=name)

plt.xticks(DATA_SIZES, labels=[f'{s:,}' for s in DATA_SIZES], rotation=0)

plt.legend(fontsize=12, loc='upper left')
plt.tight_layout()

plt.show()
```

# 6. 결과에 대한 논의 및 고찰

위 코드들을 실행하면서 각 정렬 마다 다양한 크기의 랜덤 데이터를 적용해 수행 시간을 비교할 수 있었습니다. 그 결과 크게 버블, 삽입, 선택 정렬 그룹과 병합, 퀵 정렬 그룹간의 명확한 시간 복잡도의 차이를 볼 수 있었습니다.

- 버블, 삽입, 선택 정렬 : 이 알고리즘들은 데이터의 크기가 커질수록 실행 시간이 기하급수적으로 증가했습니다. 10만개의 데이터 테스트에서 약 200초에서 600초에 가까운 시간이 소요되어, 대용량 데이터 처리에는 적합하지 않다는 점을 알 수 있습니다.

  1. 버블 정렬: 데이터가 거의 정렬되어 있어도 비교 횟수가 거의 N**2에 가까워 비효율성이 가장 높았습니다.

  2. 삽입 정렬: 데이터가 거의 정렬되어 있다면 O(N)에 가까운 성능을 내는 장점이 있어 무작위 데이터에서는 낮은 성능을 보였습니다.

  3. 선택 정렬: 교환 횟수는 O(N)으로 적지만, 비교 횟수가 O(N**2)이므로 대용량 데이터에서는 삽입 정렬과 유사하게 낮은 성능을 보였습니다.

- 병합, 퀵 정렬 : 이 알고리즘 들은 데이터의 크기가 증가하더라도 실행 시간이 1초 미만으로 매우 빠르게 증가하여, 대용량 데이터 처리에 적합하다는 점을 알 수 있습니다. 이 중 퀵정렬이 평균적으로 가장 우수한 성능을 보였고, 병합 정렬은 안정적인 성능을 보였습니다.
  
  1. 병합 정렬: 데이터의 분포에 상관없이 일관된 성능을 보이며, 합치는 과정에서 추가 메모리 공간이 필요하여 퀵 정렬에 비해 약간 느린 속도를 보였습니다.

  2. 퀵 정렬: 인플레이스(In-place) 정렬이 가능하여 메모리 효율이 좋았습니다. 위 실행 결과들을 통해 가장 빠른 수행 시간을 기록했으나, 피벗 선택이 잘못될 경우, O(N**2)으로 성능이 떨어질 위험이 있으니 주의하여 설정해야 합니다.
