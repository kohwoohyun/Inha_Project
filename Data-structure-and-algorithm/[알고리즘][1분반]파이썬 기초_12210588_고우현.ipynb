## 2. 컴퓨터의 계산

컴퓨터에게 계산을 시키는 것은 수학과 거의 동일합니다.

특이한 점은 나누기의 종류가 총 3가지로 몫만 알려주기, 나머지만 알려주기, 몫과 나머지 모두 알려주기가 있습니다.

```python
print(10 + 5)
# 더하기

print(10 - 5)
# 빼기

print(10 * 5)
# 곱하기

print(10 / 5)
# 나누기, 소수점까지 계산

print(10 / 3)
# 10 나누기 3의 정확한 결과

print(10 // 3)
# 10 나누기 3의 몫

print(10 % 3)
# 10 나누기 3의 나머지
```

## 3. 계산의 순서

컴퓨터는 우리와 똑같은 수학 규칙을 따릅니다.

곱하기와 나누기를 먼저 수행하고, 더하기와 빼기를 수행합니다.

계산의 순서를 직접 정해주고 싶을 때는 괄호()를 사용합니다.

```python
print(3 + 4 * 2)
# 컴퓨터가 알아서 순서대로 계산, 결과 : 11

print((3 + 4) * 2)
# 괄호를 사용해서 계산 순서를 정한 후 계산, 결과 : 14
```

## 4. 수학 도우미 math 라이브러리

## 2. 문자열 자유자재로 다루기

문자열은 단순히 출려만 하는 것이 아니라, 여러 가지 방법으로 가공할 수 있습니다.

문자열을 다루기 전에 컴퓨터는 숫자와 문자열을 다른 방식으로 기억합니다.

- 숫자 (int, float): 숫자 123은 호텔의 방 하나에 통째로 쏙 들어갑니다. [ 123 ]

- 문자열 (str): 문자열 "PYTHON"은 글자 하나하나가 쪼개져서 각각의 방에 차례대로 들어갑니다. 즉, 글자의 개수만큼 방을 여러 개 차지하는 셈이죠.
 [ 'P' ] [ 'Y' ] [ 'T' ] [ 'H' ] [ 'O' ] [ 'N' ]

이 차이로 문자열은 숫자와 달리 특별한 방식으로 다룰 수 있습니다.각각 번호(인덱스)가 매겨져있기 때문에 특정 위치에 있는 글자 하나만 집어내거나, 몇 번 부터 몇 번까지의 글자들을 잘라낼 수 있습니다.

프로그래밍 언어는 의미가 명확하고 모두가 동의할 수 있는 행동만 허용됩니다.

- '+'와 '*'가 가능한 이유: '이어 붙이기'와 '반복하기'는 누가 봐도 그 의미가 명확하고 직관적입니다.

- '-' (빼기)가 불가능한 이유: 앞서 살펴본대로 의미가 너무 애매모호합니다. "바나나" - "나"의 결과가 "바"인지 "바나"인지 하나의 정답을 정할 수 없죠. 그래서 파이썬은 이런 혼란을 막고, 대신 슬라이싱이라는 훨씬 더 명확한 도구를 제공합니다.

- '/' (나누기)가 불가능한 이유: 나누기 역시 마찬가지입니다. "파이썬최고" / 2의 결과가 글자 수를 반으로 나눈 "파이썬"인지, 문자열을 두 조각으로 나눈 것인지 명확한 규칙을 정할 수 없습니다.

```python
print("안녕" + "하세요")
# + 연산자, 결과: 안녕하세요

print("파이썬 최고! " * 3)
# * 연산자, # 결과: 파이썬 최고! 파이썬 최고! 파이썬 최고!
```

```python
print("Hello" - "H")
# - 연산자, 다른 방법으로 이 문제를 해결가능합니다.
```

```python
print("Hello" / "H")

print("파이썬 최고" / 3)

# / 연산자, 빼기와 마찬가지로 문자열을 나눈다는 것의 의미를 명확하게 정의할 수 없습니다.
```

### 1. 인덱싱

문자열에 들어있는 글자들은 순서대로 번호가 매겨져 있습니다. 이 번호를 인덱스(index)라고 부릅니다. 중요한 점은, 첫 번째 글자의 번호는 1이 아니라 0부터 시작한다는 점입니다.

```python
print("Python"[0])
# 결과: P

print("Python"[3])
# 결과: h

#  P  y  t  h  o  n
#  0  1  2  3  4  5 (인덱스 번호)
```

### 2. 슬라이싱

인덱스를 활용하면 문자열의 특정 부분만 잘라낼 수도 있습니다.

[시작번호:끝번호] 형식으로 사용하는데, 끝번호에 해당하는 글자는 포함되지 않는다는 점을 유의해야 합니다.

```python
text = "Life is short"
print(text[0:4])
# 0번부터 4번 '앞'까지 잘라줘! -> 결과: Life

print(text[5:7])
# 5번부터 7번 '앞'까지 잘라줘! -> 결과: is

print(text[8:])
# 8번부터 끝까지 잘라줘! -> 결과: short

print(text[:4])
# 처음부터 4번 '앞'까지 잘라줘! -> 결과: Life
```

### 3. 함수(메소드)

인덱싱이나 슬라이싱 외에도, 문자열은 자체적으로 아주 편리한 기능들을 가지고 있습니다.

함수(메소드)란 문자열이 기본적으로 가지고 있는 편리한 기능(명령어)들입니다. 문자열 데이터라면 언제든지 점(.)을 찍어서 이 기능들을 불러와 사용할 수 있습니다.

글자 수 세기: len()
- 문자열의 총 길이를 알려줍니다. (공백 포함)

대문자로 바꾸기: .upper()
- 모든 알파벳을 대문자로 바꿔줍니다.

소문자로 바꾸기: .lower()
- 모든 알파벳을 소문자로 바꿔줍니다.

글자 바꾸기: .replace()
- replace("바꿀글자", "새글자") 형식으로 사용합니다.

글자 위치 찾기: .find()
- 특정 글자가 몇 번째 인덱스에 처음 나타나는지 알려줍니다. 만약 글자가 없으면 -1을 알려줍니다.

```python
text = "Hello, Python!"

print("글자 수:", len(text))
# 글자 수를 보여줍니다.

print("대문자로:", text.upper())
# 문자열을 대문자로 바꿔줍니다.

print("소문자로:", text.lower())
# 문자열을 소문자로 바꿔줍니다.

print("글자 바꾸기:", text.replace("!", "?"))
# '!'을 '?'로 바꿔줍니다.

print("'P'의 위치:", text.find("P"))
# 'P'의 위치를 알려줍니다.

print("'z'의 위치:", text.find("z"))
# 없는 글자라 -1이 나와요.
```

```python
# 사용자가 아이디와 비밀번호를 입력했다고 상상해봅시다.
input_id = "  PythonMaster  "
input_pw = "password123"

print("===== 아이디 검사 시작 =====")

# 1. 불필요한 공백 제거하기 (.replace)
cleaned_id = input_id.replace(" ", "")
print("공백 제거:", cleaned_id)
# 사용자가 실수로 입력한 앞뒤 공백을 없애줍니다.

# 2. 아이디 길이 제한하기 (len)
print("아이디 길이:", len(cleaned_id))
# 아이디가 5글자 이상, 15글자 이하인지 확인합니다.

# 3. 아이디를 소문자로 통일하기 (.lower)
final_id = cleaned_id.lower()
print("최종 아이디:", final_id)
# 'Python'과 'python'을 같은 아이디로 취급하기 위해 모두 소문자로 바꿉니다.

print("\n===== 비밀번호 검사 시작 =====")

# 4. 비밀번호에 특수문자 '!' 포함 여부 확인하기 (.find)
print("'!' 포함 여부 (위치):", input_pw.find("!"))
# find의 결과가 -1이 아니면, 특수문자가 포함되어 있다는 뜻입니다.
```

## 4. 데이터 타입

지금까지 정수(int), 실수(float), 그리고 문자열(str) 총 세 가지 종류의 데이터를 배웠습니다. 이렇게 데이터의 종류를 자료형 또는 데이터 타입(Data Type)이라고 부릅니다.

컴퓨터는 데이터의 종류에 따라 처리하는 방식이 완전히 다릅니다. 예를 들어, 숫자 5는 계산을 위한 데이터지만, 문자 "5"는 그저 '5'라는 모양의 글자일 뿐이죠. 이 둘을 구분하지 못하면 컴퓨터는 엉뚱한 결과를 내놓거나 오류를 일으킬 거예요.

```python
print(5 + 5)
# 숫자 5와 5를 더함 -> 결과: 10

print("5" + "5")
# 문자 "5"와 "5"를 이어 붙임 -> 결과: 55

# print(5 + "5")
# 숫자와 문자는 직접 더할 수 없어서 오류가 발생해요
```

```python
print(type(10))
# 결과: <class 'int'>

print(type(3.14))
# 결과: <class 'float'>

print(type("안녕"))
# 결과: <class 'str'>
```

때로는 문자열을 숫자로, 또는 숫자를 문자열로 바꿔야 할 때가 있습니다. 이것을 형 변환이라고 합니다.
- int(): 다른 자료형을 정수로 바꿔줘요.
- float(): 다른 자료형을 실수로 바꿔줘요.
- str(): 다른 자료형을 문자열로 바꿔줘요.

```python
num_str = "100"
num_int = int(num_str)
# 문자열 "100"을 정수 100으로 변신
print(num_int + 50)
# 이제 계산이 가능해요 -> 결과: 150

age = 20
print("제 나이는 " + str(age) + "살입니다.")
# 숫자 20을 문자열 "20"으로 변신시켜 문장과 합치기
```

- 가독성: 3.14 * 5 * 5 보다 3.14 * radius * radius 가 '반지름을 이용해 넓이를 구하는구나' 하고 훨씬 이해하기 쉬워요.

- - 변수 이름은 그 안에 어떤 데이터가 들어있는지 알 수 있도록 의미 있는 단어로 짓는 것이 좋습니다. 예를 들어 a = 100 보다는 score = 100 이 훨씬 이해하기 쉽겠죠? 파이썬에서는 여러 단어를 조합할 때 user_name 처럼 **밑줄(_)**로 연결하는 방식(snake_case)을 주로 사용합니다.

컴퓨터에게 질문을 던질 때 사용하는 기호를 비교 연산자라고 합니다.

| 연산자 | 의미              | 예시 (결과)        |
|--------|-------------------|--------------------|
| >      | 크다              | 10 > 5 (True)      |
| <      | 작다              | 10 < 5 (False)     |
| >=     | 크거나 같다 (이상) | 10 >= 10 (True)    |
| <=     | 작거나 같다 (이하) | 10 <= 5 (False)    |
| ==     | 같다              | 10 == 10 (True)    |
| !=     | 같지 않다         | 10 != 5 (True)     |

- - my_age = 20 → my_age라는 변수에 20을 저장해라.
- == (비교 연산자): "왼쪽 데이터와 오른쪽 데이터가 같니?" 라는 질문입니다.
- - my_age == 20 → my_age에 들어있는 값이 20이랑 같니?

```python
# 질문: 10은 5보다 큰가?
print(10 > 5)

# 질문: "hello"와 "world"는 같은가?
print("hello" == "world")

# 변수를 사용한 질문
my_age = 17
is_adult = (my_age >= 20)
# 17 >= 20 의 결과를 is_adult 변수에 저장

print(f"나는 성인인가요? {is_adult}")

# 결과값의 데이터 타입 확인하기
print(type(is_adult))
# 결과: <class 'bool'>
```

# 5. 조건문

## 1. 조건문의 기본, if

if는 영어 단어 뜻 그대로 "만약 ~라면" 이라는 의미를 가집니다. 컴퓨터에게 "만약 이 조건이 맞다면(True라면), 이 코드를 실행해!" 라고 알려주는 역할을 합니다.

if문을 사용할 때는 두 가지를 반드시 기억해야 합니다.

1. 콜론(:): 조건문 맨 끝에는 항상 콜론을 붙여서 "이제 조건에 따른 실행 코드가 시작됩니다!" 라는 신호를 줘야 합니다.

2. 들여쓰기(Indentation): if문 아래에 오는 실행 코드는 반드시 똑같은 간격으로 들여쓰기를 해야 합니다. (보통 스페이스바 4칸 또는 Tab키 1번) 이 들여쓰기는 "이 코드는 if문에 소속된 코드입니다" 라는 의미를 갖는 아주 중요한 약속입니다.



```python
my_age = 20

# 질문: my_age가 19보다 큰가? -> True!
if my_age > 19:
    # 위의 조건이 True이므로, 아래 코드가 실행됩니다.
    print("성인 영화를 관람할 수 있습니다.")

print("프로그램이 종료되었습니다.")
# 이 코드는 if문과 상관없이 항상 실행됩니다.
```

## 3. 더 많은 선택지, elif

### 중첩 if-else

확인해야 할 조건이 여러 개인데 우리에게 if와 else밖에 없다면 어떻게 해야 할까요? else 안에 또 다른 if-else를 넣는 방식으로 해결해야 합니다.

예제:

- score 라는 변수에 점수를 저장합니다. 만약 score 가 90 이상이면 “A학점 입니다” 80점 이상 90점 미만이면 “B학점 입니다.”, 70점 이상 80점 미만이면 “C학점 입니다.” 70점 미만이면 “재수강 입니다” 라는 결과를 파이썬 코드로 작성하세요.

- - score 라는 변수에 85를 저장한다.

- - 만약 score가 90보다 크거나 같다면,

- - "A 학점입니다!"를 화면에 출력한다.

- - 그렇지 않다면,

- - 다시 만약 score가 80보다 크거나 같다면,

- - "B 학점입니다!"를 화면에 출력한다.

- - 그렇지 않다면,

- - 다시 만약 score가 70보다 크거나 같다면,

- - "C 학점입니다!"를 화면에 출력한다.

- - 그렇지 않다면 (위의 모든 조건에 해당하지 않는다면),

- - "재수강이 필요합니다."를 화면에 출력한다.

```python
score = 85

if score >= 90:
    print("A 학점입니다!")
else:
    # 90점 미만인 경우, 다시 80점 이상인지 확인
    if score >= 80:
        print("B 학점입니다!")
    else:
        # 80점 미만인 경우, 다시 70점 이상인지 확인
        if score >= 70:
            print("C 학점입니다!")
        else:
            # 모든 조건에 해당하지 않는 경우
            print("재수강이 필요합니다.")
```

문제의 복잡성을 편리하게 해결하기 위해서 elif가 등장했습니다.

if-elif-else 구조에서는 위에서부터 순서대로 조건을 확인하다가, True인 조건을 하나라도 만나면 그 코드만 실행하고 나머지 elif나 else는 확인조차 하지 않고 바로 빠져나온다는 점이 중요합니다.


```python
score = 85

if score >= 90:
    print("A 학점입니다!")
elif score >= 80:
    # 85는 90보다 크지 않으니 첫 if는 통과.
    # 85는 80보다 크거나 같으니 여기서 True! 아래 코드를 실행하고 조건문 전체를 빠져나갑니다.
    print("B 학점입니다!")
elif score >= 70:
    print("C 학점입니다!")
else:
    print("재수강이 필요합니다.")
```

## 4. 조건문 기본 구조

if 첫 번째 조건:

    첫 번째 조건이 True일 때 실행할 코드
elif 두 번째 조건:

    첫 번째 조건은 False이고, 두 번째 조건이 True일 때 실행할 코드
elif 세 번째 조건:

    ...
else:

    위의 모든 조건이 False일 때 실행할 코드

## 3. 리스트는 문자열의 형님?

| 구분   | 문자열 (str)                                | 리스트 (list)                                   |
|--------|---------------------------------------------|------------------------------------------------|
| 공통점 | 순서(sequence)가 있는 데이터의 묶음          | 순서(sequence)가 있는 데이터의 묶음              |
| 인덱싱 | `str[0]` -> 한 글자                         | `list[0]` -> 한 요소(element)                   |
| 슬라이싱 | `str[1:3]` -> 여러 글자                     | `list[1:3]` -> 여러 요소                        |
| 차이점 | 한 번 만들면 내용을 바꿀 수 없음 (immutable) | 내용을 마음대로 바꾸고 추가할 수 있음 (mutable) |
| 내용물 | 오직 글자만 담을 수 있음                     | 숫자, 문자열, 심지어 다른 리스트까지 무엇이든 담을 수 있음 |

```python
my_list = [10, "hello", True, 3.14]

# 인덱싱
print(my_list[0]) # 결과: 10
print(my_list[1]) # 결과: "hello"

# 슬라이싱
print(my_list[1:3]) # 결과: ["hello", True]

# 리스트는 수정이 가능
my_list[0] = 99
print(my_list) # 결과: [99, 'hello', True, 3.14]
```

# 9. 반복문

우리는 컴퓨터에게 "이 리스트에 데이터가 몇 개가 있든, 그냥 처음부터 끝까지 알아서 하나씩 전부 확인해!" 라고 똑똑하게 지시할 방법이 필요합니다.
이 '자동으로 처음부터 끝까지'라는 작업을 수행해주는 것이 바로 반복문 (Loop) 입니다.

## 1. 첫 번째 반복문: while

while은 "~하는 동안에 계속해!" 라는 뜻으로, 주어진 조건이 True인 동안에만 코드를 계속해서 반복 실행합니다.

while문을 사용하려면 보통 세 가지가 필요합니다.
1. 시작 조건 (초기화): 반복을 시작할 숫자 변수를 하나 만듭니다. (예: i = 0)
2. 반복 조건(종료 조건): while 뒤에 언제까지 반복할 것인지 조건을 적습니다. (예: while i < 5)
3. 조건 변경(증감 조건): 반복문이 끝날 때마다 시작 조건으로 만들었던 변수의 값을 바꿔줍니다. (예: i = i + 1) 이 부분이 없으면 무한히 반복하는 무한 루프에 빠지니 정말 중요해요!

### 추가 예제

예제 1: 회원가입 프로그램 만들기

## 2. 더 똑똑한 반복문: for

for문은 "리스트(컨테이너)에 있는 모든 데이터를 하나씩 꺼내면서 반복" 하는 반복문입니다. while문처럼 시작 조건을 만들거나, 직접 값을 증가시킬 필요가 없습니다.

```python
# while문
i = 0
while i < 5:
    print("안녕하세요^^")
    i += 1

# for문
for i in range(5):
    print("안녕하세요^^")
```

## 3. 반복문의 흐름 제어하기: break, continue, pass

때로는 반복문을 돌다가 중간에 멈추거나, 특정 순서를 건너뛰어야 할 때가 있습니다.
- break (탈출): break를 만나면, 반복 횟수가 남았더라도 그 즉시 반복문 전체를 완전히 빠져나옵니다. 아이디 중복 검사 예제처럼, 원하는 결과를 이미 찾았을 때 더 이상 불필요한 반복을 하지 않기 위해 사용됩니다.

- continue (다음으로): continue를 만나면, 이번 반복만 여기까지만 하고, 다음 순서로 바로 넘어갑니다. 반복문 전체를 빠져나가지 않는다는 점에서 break와 다릅니다.

- pass (일단 통과): pass는 말 그대로 아무것도 하지 않고 그냥 지나가는 역할입니다. "나중에 여기에 코드를 채워 넣을 건데, 지금 당장 문법 오류는 나지 않게 해줘!" 라는 의미의 임시 표식으로 사용됩니다.

### 추가 예제

```python
# 리스트에서 첫 번째 짝수 찾기
# break

numbers = [1, 3, 5, 7, 8, 9, 11]
for num in numbers:
    print(f"현재 확인 중인 숫자: {num}")
    if num % 2 == 0:
        print(f"찾았다! 첫 번째 짝수는 {num}입니다.")
        break # 짝수를 찾았으니 더 이상 반복할 필요가 없습니다.

# 특정 유저를 제외하고 환영 메시지 보내기
# continue

user_ids = ['Frodo', 'Neo', 'Tube', 'Ryan', 'Muzi']
banned_user = "Tube"
for user in user_ids:
    if user == banned_user:
        continue # Tube는 건너뛰고 다음 사람으로 넘어갑다.
    print(f"안녕하세요 {user}님, 환영합니다!")

# 기능 개발 준비
# pass

user_ids = ['Frodo', 'Neo', 'admin', 'Ryan']
for user in user_ids:
    if user == 'admin':
        pass # 일단 아무것도 안하고 넘어간다.
    else:
        print(f"{user}님에게 일반 광고 메일을 보냅니다.")
```

## 4. 프로그래밍의 발전 과정: in 연산자의 정체

in 연산자는 for문, if문, break를 사용해 길게 만들었던 코드의 최종 완성형 버전입니다.

프로그래밍 언어는 다음과 같은 순서로 발전해 온 셈입니다.

2. 수동 반복 (while): 가능하지만, 개발자가 직접 제어할 게 많아 실수하기 쉬움.
3. 자동 반복 (for): 가장 표준적이고 안전하며 가독성 좋은 반복 방식.
4. 추상화된 반복 (in): '포함 여부 확인'처럼 특정 목적에 최적화된 가장 편리한 방식

# 10. 함수

코딩을 하다보면 다음과 같은 문제점들이 발생할 수 있습니다.

1. 코드 낭비: 똑같은 역할을 하는 코드가 중복으로 쓰이고 있습니다.
2. 유지보수 최악: 만약 로직을 바꿔야 한다면? 해당 코드들을 모두 찾아서 똑같이 수정해야 합니다. 하나라도 빼먹으면 큰 오류로 이어질 겁니다.

자주 쓰는 기능을, 하나의 '코드 레시피'처럼 만들어두고 필요할 때마다 이름만 불러서 재사용하도록 하는 것이 바로 함수(Function)입니다.

## 1. 함수 만들고 사용하기

함수는 특정 작업을 수행하는 코드 덩어리에 이름을 붙여둔 것입니다.

1. 함수 정의(Define): def 키워드를 사용해 '레시피'를 만듭니다.
2. 함수 호출(Call): 만들어둔 함수의 이름을 불러 '레시피'를 사용합니다.

### 추가 예제
아이디 중복 확인 기능을 check_id_exists라는 이름의 함수로 만들어 봅니다.

```python
user_ids = ['Frodo', 'Neo', 'Tube', 'Ryan', 'Muzi']

# 함수 정의
# id_to_check 라는 매개변수를 받아서 아이디 존재 여부를 확인하는 함수
def check_id_exists(id_to_check):
    is_duplicate = False
    for existing_id in user_ids:
        if id_to_check == existing_id:
            is_duplicate = True
            break

    # 여기서 바로 결과를 출력
    if is_duplicate:
        print("이미 사용 중인 아이디입니다.")
    else:
        print("존재하지 않는 아이디입니다.")

# 함수 호출
print("--- 회원가입 중 ---")
check_id_exists("Ryan") # "Ryan"이라는 매개변수를 넣어서 함수 실행

print("\n--- 로그인 중 ---")
check_id_exists("JayG") # "JayG"라는 매개변수를 넣어서 함수 실행
```

## 2. 함수의 핵심 return - 발표 vs 보고

위의 예제의 함수는 아이디가 있는지 없는지 결과를 화면에 발표할 뿐, 결과를 알려주지는 않습니다.

- print (단순 발표): "여러분, 아이디가 중복입니다!" 라고 스피커로 외치는 것과 같습니다. 프로그램의 다른 부분(예: if문)은 이 발표를 듣고 다음 행동을 결정할 수 없습니다.
- return (결과 보고): 함수가 조용히 호출한 곳으로 찾아와 "결과는 True(중복 맞음)입니다." 라고 보고하는 것과 같습니다. 프로그램은 이 보고 결과를 변수에 저장하거나 if문에 바로 사용할 수 있습니다.

### 추가 예제

위의 함수에서 결과를 return하고, 그 값을 받은 쪽에서 상황에 맞게 print를 합니다.

```python
# return을 사용하는 함수
def check_id_exists_v2(id_to_check):
    is_duplicate = False
    for existing_id in user_ids:
        if id_to_check == existing_id:
            is_duplicate = True
            break
    return is_duplicate # 결과를 True 또는 False로 보고

# return 값을 활용하는 프로그램

# 1. 회원가입 기능
new_id = "Ryan"
# 함수에게 "Ryan"의 존재 여부를 물어보고, 그 결과를 result 변수에 저장
result = check_id_exists_v2(new_id)
if result == True:
    print(f"{new_id}는 이미 사용 중이라 가입할 수 없습니다.")
else:
    print(f"{new_id}는 가입 가능합니다.")


# 2. 로그인 기능
login_id = "JayG"
# if문에 바로 함수의 보고 결과를 사용
if check_id_exists_v2(login_id) == True:
    print(f"{login_id}님, 비밀번호를 입력하세요.")
else:
    print(f"{login_id}는 존재하지 않는 아이디입니다.")

# 추가
# 3. 회원 탈퇴 기능
delete_id = input("탈퇴할 아이디를 입력하세요: ")

if check_id_exists_v2(delete_id):  # 함수 재사용
    user_ids.remove(delete_id)     # 리스트에서 삭제
    print(f"{delete_id} 아이디가 탈퇴되었습니다.")
else:
    print("존재하지 않는 아이디입니다.")
```

## 4. 함수, 그 다음은? (클래스 맛보기)

'행동'은 완벽하게 똑같지만, 확인해야 할 '데이터'가 다르다는 이유만으로 비슷한 함수 두 개 이상을 만들어야 하는 상황이 생길 수 있습니다. 이를 해결하기 위해 클래스(Class)와 객체(Object)를 이용합니다.

클래스: 데이터와 함수를 품은 '설계도'

클래스는 연관 있는 데이터(변수)와 그 데이터를 다루는 함수를 하나로 묶어놓은 '설계도'와 같습니다.

- 데이터 (속성, Attribute): 회원 목록 (user_ids)
- 함수 (메소드, Method): 아이디 확인 기능 (check_id_exists)

이 설계도를 사용해 실제 메모리에 만들어진 실체를 객체(Object) 또는 인스턴스(Instance)라고 부릅니다.

### 추가 예제

클래스를 사용한 아이디 확인기 설계도

```python
# '아이디 확인기' 라는 설계도 생성
class IdChecker:
    # 생성자: 객체를 만들 때 회원 목록(user_list)을 재료로 받는다.
    def __init__(self, user_list):
        # 받은 회원 목록을 이 객체만의 데이터로 저장한다.
        self.users = user_list

    # 메소드: 이 객체가 가진 회원 목록(self.users)을 확인하는 기능
    def check_id_exists(self, id_to_check):
        is_duplicate = False
        for existing_id in self.users:
            if id_to_check == existing_id:
                is_duplicate = True
                break
        return is_duplicate

# 각 서비스의 회원 정보 DB
kakao_users = ['Frodo', 'Neo', 'Tube']
naver_users = ['Ryan', 'Muzi', 'Guest']

# '카카오 전용 아이디 확인기' 객체를 만든다. (재료: kakao_users)
kakao_checker = IdChecker(kakao_users)

# '네이버 전용 아이디 확인기' 객체를 만든다. (재료: naver_users)
naver_checker = IdChecker(naver_users)

# 이제 각 객체에게 일을 시킵니다.

# 카카오 확인기에게 물어보기
if kakao_checker.check_id_exists("Neo"):
    print("카카오: 'Neo'는 이미 사용 중인 아이디입니다.")

# 네이버 확인기에게 물어보기
if naver_checker.check_id_exists("Neo"):
    print("네이버: 'Neo'는 이미 사용 중인 아이디입니다.")
else:
    print("네이버: 'Neo'는 사용 가능한 아이디입니다.")
```
