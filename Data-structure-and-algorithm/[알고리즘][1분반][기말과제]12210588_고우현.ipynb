#1. 다익스트라 알고리즘

```python
processed = []

def find_lowest_cost_node(costs):
    lowest_cost = float("inf")
    lowest_cost_node = None
    for node in costs:
        cost = costs[node]
        if cost < lowest_cost and node not in processed:
            lowest_cost = cost
            lowest_cost_node = node
    return lowest_cost_node

def dijkstra(graph, start, final):
    node = find_lowest_cost_node(costs)
    while node is not None:
        cost = costs[node]
        neighbors = graph[node]
        for n in neighbors.keys():
            new_cost = cost + neighbors[n]
            if costs[n] > new_cost:
                costs[n] = new_cost
                parents[n] = node
        processed.append(node)
        node = find_lowest_cost_node(costs)

    path = []
    current_node = final
    while current_node != start:
        path.append(current_node)
        next_node = parents[current_node]
        current_node = next_node
    path.append(start)

    return costs[final], path[::-1]
```

#2. 너비우선탐색(BFS) 알고리즘

```python
from collections import deque

def bfs(graph,start,final):
    search_queue = deque()
    search_queue.append(([start],0))

    shortest_len = None
    results = []

    min_cost = float('inf')
    min_path = None

    while search_queue:

        path, cost = search_queue.popleft()
        node = path[-1]

        if shortest_len is not None and len(path) > shortest_len:
            continue

        if node == final:
            shortest_len = len(path)
            results.append((path, cost))

            if cost < min_cost:
                min_cost = cost
                min_path = path

            continue

        for adjacent in graph.get(node,[]):
            if adjacent not in path:
                new_path = path[:]
                new_path.append(adjacent)
                new_cost = cost + graph[node][adjacent]
                search_queue.append((new_path, new_cost))

    if min_path:
        print("최단 경로 :", min_path, "최소 비용 :", min_cost)

    for path, cost in results:
        print((path, cost))
```

```python
bfs(graph, 'A', 'I')
```

#3. 탐욕적 기법

```python
def greedy(graph, start, final):
    path = [start]
    cost = 0
    current = start

    while current != final:
        neighbors = graph.get(current, {})

        if not neighbors:
            print("더 이상 이동할 수 없습니다.")
            return

        next_node = None
        min_weight = float('inf')

        for adjacent in neighbors:
            if adjacent not in path:
                weight = neighbors[adjacent]
                if weight < min_weight:
                    min_weight = weight
                    next_node = adjacent

        if next_node is None:
            print("탐욕적으로 선택할 수 있는 경로가 없습니다.")
            return

        path.append(next_node)
        cost += min_weight
        current = next_node

    print("경로 :", path, "해당 경로의 비용 :", cost)
```

```python
greedy(graph, 'A', 'I')
```

#4. 배낭 채우기 문제

```python
weight_of_items = {'A':2,'B':3,'C':5,'D':7,'E':1,'F':4,'G':1}
```

```python
weight_of_items
```

```python
value_of_items = [('A',10),('B',5),('C',15),('D',7),('E',6),('F',18),('G',3)]
```

```python
value_of_items
```

```python
W = 15
```

```python
def greedy_knapsack(W, wt,val):
    knapsack = []
    weight = 0
    value = 0

    while weight < W:
        selected = max(val, key=lambda item:item[1])
        item = selected[0]
        i = val.index(selected)

        if wt[item] <= W - weight:
            knapsack.append(selected[0])
            weight += wt[item]
            value += selected[1]
        del val[i]

        if len(val) == 0:
            break

    return knapsack, weight, value
```

```python
greedy_knapsack(W, weight_of_items, value_of_items)
```

```python
def dynamic_knapsack(W,wt,val,n):

    K = [[0 for x in range(W+1)] for x in range(n+1)]

    for i in range(n+1):
        for w in range(W+1):
            if i==0 or w==0:
                K[i][w] = 0
            elif wt[i-1] <= w:
                K[i][w] = max(val[i-1] + K[i-1][w-wt[i-1]], K[i-1][w])
            else:
                K[i][w] = K[i-1][w]

    return K[n][W]
```

```python
val = [10, 5, 15, 7, 6, 18, 3]
wt = [2, 3, 5, 7, 1, 4, 1]
W = 15
n = len(val)
```

```python
dynamic_knapsack(W,wt,val,n)
```

##결과 논의

greedy_knapsack과 dynamick_knapsack 모두 동일한 가치 54를 결과값으로 나왔습니다.

이는 greedy 알고리즘이 가치가 가장 큰 물건부터 선택하는 기준으로 실행해 선택 순서가 우연히 최적해와 일치했기 때문입니다.

그러나 탐욕적 기법은 이번 경우에는 최적해를 찾았지만 지역 최적을 선택하고 항상 전역 최적해를 보장하지 않고, 동적계획법은 모든 경우의 수를 고려해 항상 전역 최적해를 보장하므로 빠르게 답을 찾기 위해서는 탐욕적 기법을, 느리지만 확실한 전역 최적해를 찾기 위해서는 동적계획법을 활용해야 합니다.

```python
def dynamic_knapsack(W, wt, val, n):

    K = [[0 for x in range(W+1)] for x in range(n+1)]

    for i in range(n+1):
        for w in range(W+1):
            if i==0 or w==0:
                K[i][w] = 0
            elif wt[i-1] <= w:
                K[i][w] = max(val[i-1] + K[i-1][w-wt[i-1]], K[i-1][w])
            else:
                K[i][w] = K[i-1][w]

    selected_items = []
    w = W

    for i in range(n, 0, -1):
        if K[i][w] != K[i-1][w]:
            selected_items.append(i-1)
            w -= wt[i-1]

    selected_items.reverse()

    total_weight = sum(wt[i] for i in selected_items)
    total_value = K[n][W]

    return selected_items, total_weight, total_value
```

```python
dynamic_knapsack(W,wt,val,n)
```

#5. 배송과 물류 관점

너비 우선 탐색은 중간 물류 거점을 가장 적게 거치는 지역부터 배송을 목표로 합니다. 즉, 환적 횟수가 적은 지역부터 배달 순서를 정하는 알고리즘입니다.

다익스트라 알고리즘은 배송 비용이나 이동 시간을 최소화하여 총 배송 비용이 최소가 되는 경로를 찾는 알고리즘입니다.

탐욕적 기법은 트럭의 적재 한도가 존재할 때, 매 순간 가장 가치가 큰 화물이나 단위 무게당 수익이 높은 화물부터 트럭에 적재하는 등의 트럭 안에 택배를 어떻게 실을 것인가를 결정하는 알고리즘입니다.

동적계획법은 기사의 수, 근무 시간, 지역 물량, 배송 제한 조건 등의 다양한 요소들을 모두 고려하여 전체 배송 효율이 최대가 되도록 하는 기사-지역 배정을 수행하는 알고리즘입니다. 따라서 모든 경우를 체계적으로 비교하여 최적 배정을 계산합니다.

따라서 네 알고리즘은 모두 배송 및 물류 시스템에서 서로 다른 의사결정 단계를 담당하고 있습니다.
