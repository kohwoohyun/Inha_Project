```python
import pandas as pd
import numpy as np
from docplex.mp.model import Model
```

```python
file_path1 = "C:/Users/kohwo/차 종류 1일 이익.xlsx"
file_path2 = "C:/Users/kohwo/방문객수.xlsx"

df_profit = pd.read_excel(file_path1, sheet_name='시트1')
df_visitor = pd.read_excel(file_path2, sheet_name='시트1')
```

```python
# -------------------------------
# 1) 인덱스 설정
# -------------------------------

I = sorted(df_profit['차급'].unique())

B = sorted(df_profit['차종'].unique())

J = sorted(df_profit['유종'].unique())

K = sorted(df_visitor['그룹유형_k'].unique())

print(I)

print(B)

print(J)

print(K)
```

```python
# -------------------------------
# 2) 파라미터 생성
# -------------------------------

# -------------------------------
# 🔹 P_bij 생성 (그룹 k는 포함하지 않음)
# -------------------------------
P_bij = {}
for _, row in df_profit.iterrows():
    b = row['차종']
    i = row['차급']
    j = row['유종']
    P_bij[(b, i, j)] = row['1일 표준대여료(원)']

# -------------------------------
# c_bik 생성 (정원 충족 여부 조건 기반)
# -------------------------------
# 그룹 k를 수용하려면 차량 탑승인원 >= 그룹 최소 정원 조건
c_bik = {}

for _, car in df_profit.iterrows():
    b = car['차종']           # 브랜드
    i = car['차급']           # 차급
    capacity = car['탑승인원']  # 차량 탑승 정원

    for _, group in df_visitor.iterrows():
        k = group['그룹유형_k']           # 그룹 유형
        min_required = group['최소_필요_정원']  # 해당 그룹의 최소 필요한 정원

        # 조건 충족 시: 최소 필요 정원 값을 넣음
        if capacity >= min_required:
            c_bik[(b, i, k)] = min_required
        else:
            c_bik[(b, i, k)] = 0


# -------------------------------
# D_k 생성 (그룹별 수요: 일일 방문수)
# -------------------------------
D_k = { row['그룹유형_k']: row['일일_방문수'] for _, row in df_visitor.iterrows() }

# -------------------------------
# a_bij 생성 (차량 최대 보유대수 사용)
# -------------------------------
a_bij = {}
for _, row in df_profit.iterrows():
    b = row['차종']
    i = row['차급']
    j = row['유종']
    a_bij[(b,i,j)] = row['최대보유대수(대)']

print("P sample:", list(P_bij.items())[:5])
print("c sample:", list(c_bik.items())[:5])
print("D:", D_k)
print("a sample:", list(a_bij.items())[:5])
```

```python
# -------------------------------
# 허용 가능한 집합 생성
# -------------------------------

def build_A(I, K):
    A = set()   # (i,k)만 포함

    for k in K:
        if k in ('1인(혼행)', '2인(커플)'):
            # 모든 차급 배정 가능
            for i in I:
                A.add((i, k))

        elif k == '3인(소가족)':
            # 경차 제외 (나머지는 허용)
            for i in I:
                if i != '경차':
                    A.add((i, k))

        elif k == '4인(가족)':
            # 경차, 소형SUV, 준중형, 준중형SUV 제외 → 나머지만 허용
            excluded = {'경차', '소형SUV', '준중형', '준중형SUV'}
            for i in I:
                if i not in excluded:
                    A.add((i, k))

        else:
            for i in I:
                if i == '승합':  # 승합차만 허용
                    A.add((i,k))
    return A

# 집합 생성
A = build_A(I, K)

# 출력 확인
print("허용 (i,k) 쌍 샘플:", sorted(list(A))[:12])
print("총 허용 가능한 (i,k) 조합 수:", len(A))
```

```python
# -------------------------------
# 선호 / 비선호 차량 비율 반영
# -------------------------------
preferred_type = {'경차','소형SUV','준중형','중형'}
nonpreferred_type = {'대형','대형(고급)','대형SUV','대형SUV(고급)','승합'}

L_P = 0.348  # 선호차량 최소 34.8%
U_N = 0.073  # 비선호차량 최대 7.3%

# 고급차 집합 (프리미엄 차량군)
I_luxury = {'중형(고급)', '중형SUV(고급)', '대형(고급)', '대형SUV(고급)'}

# 승합차만 따로 분리
I_van = {'승합'}

# -------------------------------
# 저가, 중가, 고가 분류
# -------------------------------
prices = list(P_bij.values())

Low_Price_Max = np.percentile(prices, 33)   # 하위 33%
Mid_Price_Max = np.percentile(prices, 66)   # 하위 66%

print("저가 상한:", Low_Price_Max)
print("중가 상한:", Mid_Price_Max)


# ------ 가격기반 혼합 집합 ------
I_lowprice  = {(b,i,j) for (b,i,j), p in P_bij.items() if p <= Low_Price_Max}    # 저가
I_midprice  = {(b,i,j) for (b,i,j), p in P_bij.items() if Low_Price_Max < p <= Mid_Price_Max}    # 중가
I_highprice = {(b,i,j) for (b,i,j), p in P_bij.items() if p > Mid_Price_Max}    # 고가

# ----------- 비율 제한 ----------
alpha_high = 0.277   # 고가 차량 최대 27.7%
alpha_mid  = 0.341   # 중가 차량 최대 34.1%
alpha_low_min = 0.382  # 저가 차량 최소 38.2%

# -------- 고급, 승합 제한 -------

beta_luxury = 0.023
beta_van = 0.0083
```

```python
# -------------------------------
# 3) 모델 생성
# -------------------------------
model = Model(name="Fleet_Assignment_NoPriority")

# -------------------------------
# 4) 의사결정변수: ㅌ[b,i,j,k]
#    단, (i,k)가 A에 포함된 경우만 생성
# -------------------------------
x = {}
for b in B:
    for i in I:
        for j in J:
            for k in K:
                if ((b,i,j) in P_bij.keys()) and ((i,k) in A):
                    x[(b,i,j,k)] = model.integer_var(lb=0, name=f"x_{b}_{i}_{j}_{k}")

# -------------------------------
# 5) 목적 함수: 총 수익 최대화
# -------------------------------
model.maximize(model.sum(P_bij[(b,i,j)] * var for (b,i,j,k), var in x.items()))

# -------------------------------
# 6) 제약식
# -------------------------------

# (1) 공급 제약: sum(c_bik * x) ≤ D_k
for k in K:
    expr = model.sum(c_bik[(b,i,k)] * var
                 for (b,i,j,k2), var in x.items() if k2 == k)
    model.add_constraint(expr <= D_k[k], ctname=f"supply_{k}")

# (2) 보유대수 제약: sum(x[b,i,j,k]) ≤ a_bij
for b in B:
    for i in I:
        for j in J:
            expr = model.sum(var for (b2,i2,j2,k), var in x.items()
                         if (b2,i2,j2) == (b,i,j))
            if (b,i,j) in a_bij:
                model.add_constraint(expr <= a_bij[(b,i,j)], 
                                 ctname=f"cap_{b}_{i}_{j}")

# (3) 최대 탑승 인원 제약

one_person_only = {'레이 (1인승 밴)'}

for b in one_person_only:
    for i in I:
        for j in J:
            for k in K:
                if k != '1인(혼행)' and (b,i,j,k) in x:
                    model.add_constraint(x[b,i,j,k] == 0)



# total T
T = model.sum(x[b,i,j,k] for (b,i,j,k) in x.keys())

# (4) 선호, 비선호 제약: sum(x[b,i,j,k] in nonpreferred) ≤ U_N * T,
#                       sum(x[b,i,j,k] in preferred) ≥ L_P * T

# nonpreferred 최대비율 U_N 적용
model.add_constraint(model.sum(x[b,i,j,k] for (b,i,j,k) in x.keys() if i in nonpreferred_type) 
                     <= U_N * T, ctname='nonpreferred_upper')


# preferred 최소비율 L_F 적용
model.add_constraint(model.sum(x[b,i,j,k] for (b,i,j,k) in x.keys() if i in preferred_type)
                     >= L_P * T, ctname='preferred_lower')

# (5) 고급, 승합 제약: sum(x[b,i,j,k] in I_luxury) ≤ beta_lusury * T,
#                     sum(x[b,i,j,k] in  I_van) ≤ beta_van * T

# 고급차 상한
model.add_constraint(model.sum(x[b,i,j,k] for (b,i,j,k) in x.keys() if i in I_luxury) 
                     <= beta_luxury * T, ctname='LuxuryCarLimit')

# 승합차 상한
model.add_constraint(model.sum(x[b,i,j,k] for (b,i,j,k) in x.keys() if i in I_van) 
                     <= beta_van * T, ctname='VanLimit')


# (6) 가격대별 제약: sum(x[b,i,j,k] in I_highprice) ≤ alpha_high * T
#                   sum(x[b,i,j,k] in I_midprice) ≤ alpha_mid * T
#                   sum(x[b,i,j,k] in I_lowprice) ≤ alpha_low_min*T

# 고가 차량 비율 제한 (전체의 최대 15%)
model.add_constraint(model.sum(x[b,i,j,k] for (b,i,j,k) in x.keys() if (b,i,j) in I_highprice)
                     <= alpha_high * model.sum(x.values()), ctname='Highprice_upper')

# 중가 차량 비율 제한 (전체의 최대 40%)
model.add_constraint(model.sum(x[b,i,j,k] for (b,i,j,k) in x.keys() if (b,i,j) in I_midprice)
                     <= alpha_mid * model.sum(x.values()), ctname='Midprice_upper')

# 저가 차량 최소 배정 비율 (전체의 최소 30%)
model.add_constraint(model.sum(x[b,i,j,k] for (b,i,j,k) in x.keys() if (b,i,j) in I_lowprice)
                     >= alpha_low_min * model.sum(x.values()), ctname='Lowprice_lower')


# -------------------------------
# 7) 모델 풀기 및 출력
# -------------------------------
sol = model.solve(log_output=True)

if sol:
    print("Objective value:", model.objective_value)
    results=[]
    for (b,i,j,k),var in x.items():
        if var.solution_value>1e-6:
            results.append({
                '차종':b, '차급':i, '유종':j, '그룹유형':k,
                '배정대수':int(round(var.solution_value))
            })
    df_result=pd.DataFrame(results)
    display(df_result)
else:
    print("No solution found.")
```
